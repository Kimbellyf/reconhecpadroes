```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def ler_csv(file, separador, encode, zebrado=False):
    return pd.read_csv(file, sep=separador, encoding=encode, squeeze = zebrado)
```

```python
df_all_base = ler_csv("./Base/Base_Completa.csv", ';', 'iso-8859-1', True)
```

```python
dataset=df_all_base
```

```python
dataset.describe()
```

```python
[x for x in dataset.columns]
```

```python

#dividir dados em atributos e rótulos
X = dataset[['NU_MATRICULADOS_CENSO_5EF', 'NU_PRESENTES_5EF']]
#y = dataset['rotulo']
y = dataset[['MEDIA_TOTAL_5EF']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
